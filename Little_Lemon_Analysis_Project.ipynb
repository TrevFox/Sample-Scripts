```python
# Task 2 - 
# Implement Peak Hours Stored Procedure
cursor.execute("DROP PROCEDURE IF EXISTS PeakHours;")

# Create PeakHours Procedure
print("Creating PeakHours Procedure")
CreatePeakHoursProc = """
CREATE PROCEDURE PeakHours()

BEGIN

SELECT HOUR(BookingSlot) as BookingHour,
        COUNT(BookingID) as Total_Bookings
FROM Bookings
GROUP BY BookingHour
ORDER BY Total_Bookings DESC
;

END

"""

# Run Create Procedure Query
cursor.execute(CreatePeakHoursProc)

# Call the stored procedure with its name
print("Calling PeakHours Procedure")
cursor.callproc("PeakHours")
print()

# Retrieve recrods in "dataset"
print("Results of PeakHours Procedure:")
results = next(cursor.stored_results())
dataset = results.fetchall()

# Retrieve column names using list comprehension in a 'for' loop 
for column_id in cursor.stored_results():
    columns = [column[0] for column in column_id.description]

# Print column names
print(columns)

# Print data 
for data in dataset:
    print(data)
```

```python
# Task 3 - 
# Implement Guest Status Stored Procedure
cursor.execute("DROP PROCEDURE IF EXISTS GuestStatus;")

# Create GuestStatus Procedure
print("Creating GuestStatus Procedure")
CreateGuestStatusProc = """
CREATE PROCEDURE GuestStatus()

BEGIN

SELECT CONCAT(BK.GuestFirstName, BK.GuestLastName) as GuestFullName,
        CASE WHEN EM.Role IN ('Manager', 'Assistant Manager')
                THEN 'Ready to pay'
            WHEN EM.Role = 'Head Chef'
                THEN 'Ready to serve'
            WHEN EM.Role = 'Assistant Chef'
                THEN 'Preparing Order'
            WHEN EM.Role = 'Head Waiter'
                THEN 'Order Served'
        END AS GuestStatus
FROM Bookings as BK
LEFT JOIN employees as EM ON BK.EmployeeID = EM.EmployeeID
;

END

"""

# Run Create Procedure Query
cursor.execute(CreateGuestStatusProc)

# Call the stored procedure with its name
print("Calling GuestStatus Procedure")
cursor.callproc("GuestStatus")
print()

# Retrieve recrods in "dataset"
print("Results of GuestStatus Procedure:")
results = next(cursor.stored_results())
dataset = results.fetchall()

# Retrieve column names using list comprehension in a 'for' loop 
for column_id in cursor.stored_results():
    columns = [column[0] for column in column_id.description]

# Print column names
print(columns)

# Print data 
for data in dataset:
    print(data)
```

```python
# Task 4 -
# Create Basic Sales Report Stored Procedure

cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport;")

# Create BasicSalesReport Procedure
print("Creating BasicSalesReport Procedure")
CreateBasicSalesReportProc = """
CREATE PROCEDURE BasicSalesReport()

BEGIN

SELECT cast(sum(BillAmount) as signed) as Total_Sales,
        cast(round(avg(BillAmount),0) as signed) as Avg_Sale,
        min(BillAmount) as Minimum_Bill_Paid,
        max(BillAmount) as Maximum_Bill_Paid
FROM Orders
;

END

"""

# Run Create Procedure Query
cursor.execute(CreateBasicSalesReportProc)

# Call the stored procedure with its name
print("Calling BasicSalesReport Procedure")
cursor.callproc("BasicSalesReport")
print()

# Retrieve recrods in "dataset"
print("Results of BasicSalesReport Procedure:")
results = next(cursor.stored_results())
dataset = results.fetchall()

# Retrieve column names using list comprehension in a 'for' loop 
for column_id in cursor.stored_results():
    columns = [column[0] for column in column_id.description]

# Print column names
print(columns)

# Print data 
for data in dataset:
    print(data)
```
